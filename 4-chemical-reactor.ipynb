```python
import matplotlib.pyplot as plt
```

```python
def plotSimulation(substanceA, substanceB, substanceC, endTime, timePoints):
    plt.clf()
    plt.title("Simulation of a Chemical Reactor")
    plt.xlabel("Time (s)")
    plt.ylabel("Quantity (gm)")
    plt.xlim(0, endTime)
    plt.plot(timePoints, substanceA, label="Substance A", color="black")
    plt.plot(timePoints, substanceB, label="Substance B", color="blue")
    plt.plot(timePoints, substanceC, label="Substance C", color="green")
    plt.legend()
    plt.grid()
    plt.pause(1)
    plt.show()
```

```python
def chemicalReactor(a, b, c, k1, k2, endTime, dt): 
    time = 0
    substanceA = []
    substanceB = []
    substanceC = []
    timePoints = []

    print("\tTIME\t \tA(I)\t \tB(I)\t \tC(I)\t")

    while (time <= endTime):
        substanceA.append(a)
        substanceB.append(b)
        substanceC.append(c)
        timePoints.append(time)
        time += dt

        a += (k2 * c - k1 * a * b) * dt  # a(dt) = a(0) + [k2 * c(0) - k1 * a(0) * b(0)] * dt
        b += (k2 * c - k1 * a * b) * dt  # b(dt) = b(0) + [k2 * c(0) - k1 * a(0) * b(0)] * dt
        c += (2 * k1 * a * b - 2 * k2 * c) * dt  # c(dt) = c(0) + [2k1 * a(0) * b(0) - 2k2 * c(0)] * dt

        print(f"\t{time:.2f}\t \t{a:.2f}\t \t{b:.2f}\t \t{c:.2f}\t")
    
    plotSimulation(substanceA, substanceB, substanceC, endTime, timePoints)
```

```python
chemicalReactor(100, 50, 0, 0.008, 0.002, 5, 0.1) # a, b, c, k1, k2, endTime, dt
```
