```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def displayCurve(controlPoints, curvePoints):
    plt.plot(controlPoints[:, 0], controlPoints[:, 1], 'ro-', label='Control Points')
    plt.plot(curvePoints[:, 0], curvePoints[:, 1], 'b-', label='Cubic Bezier Curve')
    plt.title('Cubic Bezier Curve Construction')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.grid(True)
    plt.show()
```


### Cubic Bezier Curve Formulas

#### General Formula:
- B(t) = (1 - t)^3 * P0 + 3t(1 - t)^2 * P1 + 3t^2(1 - t) * P2 + t^3 * P3

#### X and Y Coordinates:
- x(t) = (1 - t)^3 * P0x + 3t(1 - t)^2 * P1x + 3t^2(1 - t) * P2x + t^3 * P3x
- y(t) = (1 - t)^3 * P0y + 3t(1 - t)^2 * P1y + 3t^2(1 - t) * P2y + t^3 * P3y

#### Where:
- B(t) is the point on the curve at time t (0 ≤ t ≤ 1).
- P0, P1, P2, P3 are the four control points defining the curve.
- P0x, P1x, P2x, P3x are the x-coordinates of the control points.
- P0y, P1y, P2y, P3y are the y-coordinates of the control points.
- t is the parameter that controls the position along the curve (0 ≤ t ≤ 1).

```python
def cubicBezierCurve(p0, p1, p2, p3, numPoints=100):
    tValues = np.linspace(0, 1, numPoints)

    curvePoints = []
    for t in tValues:
        x = (1 - t)**3 * p0[0] + 3 * (1 - t)**2 * t * p1[0] + 3 * (1 - t) * t**2 * p2[0] + t**3 * p3[0]
        y = (1 - t)**3 * p0[1] + 3 * (1 - t)**2 * t * p1[1] + 3 * (1 - t) * t**2 * p2[1] + t**3 * p3[1]
        curvePoints.append((x, y))
    print(f"Curve Points: {curvePoints}")
    return np.array(curvePoints)
```

```python
p0 = (0, 0)
p1 = (2, 4)
p2 = (4, 0)
p3 = (6, 4)

# Construct Cubic Bezier curve
curvePoints = cubicBezierCurve(p0, p1, p2, p3)

# Display the result
displayCurve(np.array([p0, p1, p2, p3]), curvePoints)
```
